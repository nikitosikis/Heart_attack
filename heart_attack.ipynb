#Preprocesing

```python
X = data.drop('output', axis = 1)
y = data.output
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
matrix = [[0, 0, 0] for i in range(len(X_test))]
pred = pd.DataFrame(data=matrix, columns=['bins_reg', 'bins_boost', 'bins_true'])
pred.index.name = 'index'
pred.bins_true = y_test.values
pred
```

#Logistic Regression

```python
model_reg = LogisticRegression(random_state=0, max_iter=10000)
model_boost = CatBoostClassifier()
```

```python
model_reg.fit(X_train, y_train)
pred.bins_reg = model_reg.predict(X_test)
```

#Catboost

```python
model_boost.fit(X_train, y_train)
```

```python
pred.bins_boost = model_boost.predict(X_test)
```

#Accuracy

```python
pred
```

```python
(pred.bins_reg == pred.bins_boost).value_counts()
```

```python
(pred.bins_reg == pred.bins_true).value_counts()
```

```python
accuracy_score(pred.bins_true, pred.bins_reg)
```

```python
(pred.bins_boost == pred.bins_true).value_counts()
```

```python
accuracy_score(pred.bins_true, pred.bins_boost)
```
